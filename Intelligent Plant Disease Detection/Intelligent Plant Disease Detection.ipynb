## M'ODEL BUILDING

```python
model=Sequential()
#DATA AUGMENTATION
from tensorflow.keras.layers import (
    Input, Conv2D, MaxPooling2D, Dense,
    Flatten, GlobalAveragePooling2D
)

from tensorflow.keras import layers

#model.add(Input(shape=(150,150,3)))#Defining input shape first

model.add(layers.RandomFlip("horizontal"))
model.add(layers.RandomRotation(0.1))
model.add(layers.RandomZoom(0.1))

model.add(layers.RandomRotation(0.1))
model.add(layers.RandomZoom(0.1))
model.add(layers.RandomContrast(0.1))


model.add(  Conv2D(  120,   (5,5),  activation='relu')  )#120 IS FILTER COUNT ,(5*5) IS FILTER SIZE
model.add(  MaxPooling2D(  (2,2)  ,strides=(1,1)))#(2*2)IS THE MAXPOOLING MATRIX

model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2),strides=(2,2)))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2),strides=(2,2)))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2),strides=(2,2)))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2),strides=(2,2)))


#model.add(Flatten())

model.add(GlobalAveragePooling2D())#trying without flatten to fix size issur

model.add(Dense(60, activation='relu'))
model.add(Dense(30, activation='relu'))
#model.add(Dense(128, activation='relu'))
#model.add(Dense(10, activation='softmax'))
#model.add(Dense(60,input_shape=(100,100,3),activation='relu'))
#model.add(Dense(30,activation='relu'))
#model.add(Dense(30,activation='relu'))
#model.add(Dense(30,activation='relu'))
#model.add(Dense(30,activation='relu'))
#model.add(Dense(1,activation='softmax'))

model.add(Dense(1, activation='sigmoid'))#for Binnary classification

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'],run_eagerly=True)
```

```python
print(F1.shape)
print(T.shape)
```

```python
model.fit(F1,T, epochs=1, validation_split=0.15, batch_size=12)
```
